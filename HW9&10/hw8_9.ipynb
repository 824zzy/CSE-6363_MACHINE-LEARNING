## A solution

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
d1 = np.random.multivariate_normal(mean=[3, 3], cov=[[2, 0], [0, 2]], size=100)
d2 = np.random.multivariate_normal(mean=[-3, 3], cov=[[2, 0], [0, 2]], size=100)
d3 = np.random.multivariate_normal(mean=[0, -3], cov=[[2, 0], [0, 2]], size=100)
dataset = np.concatenate((d1, d2, d3))
kmeans = KMeans(n_clusters=3, random_state=1001778274)
kmeans.fit(dataset)
centers = kmeans.cluster_centers_
plt.scatter(d1[:, 0], d1[:, 1], label='Gassian Distribution1')
plt.scatter(d2[:, 0], d2[:, 1], label='Gassian Distribution2')
plt.scatter(d3[:, 0], d3[:, 1], label='Gassian Distribution3')
plt.scatter(centers[:, 0], centers[:, 1], label='K-means Clusters', marker='x', s=50)
plt.title('Sigma=2')
plt.ylabel('Y axis ')
plt.xlabel('X axis')
plt.legend()
plt.show()
print("The finded centers are:\n", kmeans.cluster_centers_)
print("The accumulate loss is: ", kmeans.inertia_)
```
